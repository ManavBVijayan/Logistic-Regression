```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data for demonstration
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = (4 + 3 * X + np.random.randn(100, 1)) > 6
```

```python
# Plot the data
plt.scatter(X, y)
plt.xlabel('Feature')
plt.ylabel('Class (1 or 0)')
plt.title('Binary Classification Data')
plt.show()
```

```python
# Logistic Regression from Scratch using Gradient Descent
X_b = np.c_[np.ones((100, 1)), X]  # Add bias term
theta = np.random.randn(2, 1)  # Random initialization of parameters

# Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Define the cost function for logistic regression
def cost_function(X, y, theta):
    m = len(y)
    h = sigmoid(X.dot(theta))
    epsilon = 1e-5  # Small value to avoid division by zero in log
    cost = -1/m * np.sum(y * np.log(h + epsilon) + (1 - y) * np.log(1 - h + epsilon))
    return cost

# Gradient Descent to optimize parameters
learning_rate = 0.01
n_iterations = 1000

for iteration in range(n_iterations):
    gradients = 1/len(y) * X_b.T.dot(sigmoid(X_b.dot(theta)) - y)
    theta = theta - learning_rate * gradients
```

```python
# Make predictions
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_proba = sigmoid(X_new_b.dot(theta))
```

```python
# Plot the decision boundary
plt.scatter(X, y)
plt.plot(X_new, y_proba, "r-", label="Decision Boundary (From Scratch)")
plt.xlabel('Feature')
plt.ylabel('Class (1 or 0)')
plt.title('Logistic Regression from Scratch')
plt.legend()
plt.show()
```
